### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
# Calcula el total de pedidos por cada producto
total_orders_per_product = order_products.groupby('product_id').size()

# Calcula el número de repeticiones de cada producto
repeated_orders_per_product = order_products[order_products['reordered'] == 1].groupby('product_id').size()

# Calcula la tasa de repetición como el número de repeticiones dividido por el total de pedidos
repetition_rate = repeated_orders_per_product / total_orders_per_product

# Muestra la tasa de repetición de los productos
repetition_rate = repetition_rate.sort_values(ascending=False)

# Imprime las tasas de repetición para los productos
print(repetition_rate.head(20))
```

Escribe aquí tus conclusiones El cálculo de la tasa de repetición de pedido por producto revela la proporción de clientes que vuelven a pedir un mismo artículo en relación con el número total de pedidos de ese producto. Esto ayuda a entender qué productos tienen una alta recurrencia y lealtad de los clientes. Los productos con una tasa de repetición alta indican que los clientes están satisfechos o interesados en comprarlos nuevamente, lo cual puede ser útil para diseñar estrategias de marketing, mejorar la experiencia de compra y promover la fidelización.
